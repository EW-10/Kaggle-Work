```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Read the data
X_full = pd.read_csv('../input/train.csv', index_col='Id')
X_test_full = pd.read_csv('../input/test.csv', index_col='Id')

# Obtain target and predictors
y = X_full.SalePrice
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = X_full[features].copy()
X_test = X_test_full[features].copy()

# Break off validation set from training data
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,
                                                      random_state=0)
```

The next code cell defines five different random forest models.  Run this code cell without changes.  (_To review **random forests**, look [here](https://www.kaggle.com/dansbecker/random-forests)._)

```python
from sklearn.ensemble import RandomForestRegressor

# Please note:
# The original notebook had models 1-5.
# I added models 6-10 below.

# Define the models
model_1 = RandomForestRegressor(n_estimators=50, random_state=0)
model_2 = RandomForestRegressor(n_estimators=100, random_state=0)
model_3 = RandomForestRegressor(n_estimators=100, criterion='absolute_error', random_state=0)
model_4 = RandomForestRegressor(n_estimators=200, min_samples_split=20, random_state=0)
model_5 = RandomForestRegressor(n_estimators=100, max_depth=7, random_state=0)

model_6 = RandomForestRegressor(n_estimators = 100, criterion = 'absolute_error', random_state = 0,\
                               max_depth=7)
model_7 = RandomForestRegressor(n_estimators = 100, criterion = 'absolute_error', random_state = 0,\
                                max_depth=9)
model_8 = RandomForestRegressor(n_estimators = 100, criterion = 'absolute_error', random_state = 0,\
                                max_depth=11)
model_9 = RandomForestRegressor(n_estimators = 100, criterion = 'absolute_error', random_state = 0,\
                                max_depth=13)
model_10 = RandomForestRegressor(n_estimators = 100, criterion = 'absolute_error', random_state = 0,\
                                max_depth=15)


models = [model_1, model_2, \
          model_3, model_4, model_5, \
          model_6, model_7, model_8, \
          model_9, model_10]
```

To select the best model out of the five, we define a function `score_model()` below.  This function returns the mean absolute error (MAE) from the validation set.  Recall that the best model will obtain the lowest MAE.  (_To review **mean absolute error**, look [here](https://www.kaggle.com/dansbecker/model-validation).)_

Run the code cell without changes.

```python
from sklearn.metrics import mean_absolute_error

# Function for comparing different models
def score_model(model, X_t=X_train, X_v=X_valid, y_t=y_train, y_v=y_valid):
    model.fit(X_t, y_t)
    preds = model.predict(X_v)
    return mean_absolute_error(y_v, preds)

# The code below was modified in order to present the original models' names:

def namestr(obj, namespace):
    return [name for name in namespace if namespace[name] is obj]

maes = {}

for i in range(0, len(models)):
    mae = score_model(models[i])
    a = namestr(models[i], globals())[0]
    a = a.capitalize()
    a = a.replace("_", " ")
    maes[a] = mae
    print(a +" MAE: %d" % (mae))
    
# The code above, used to present the original models' names
# was taken from an answer to a question on stackoverflow in the link below:  
# https://stackoverflow.com/questions/592746/how-can-you-print-a-variable-name-in-python.  
# This answer was given by jfs on Feb 26, 2009, and further edited by Community bot
# on Jun 20, 2020.
```

```python
# Creation of a dataframe with:
# (1) the model's name
# (2) it's mae 
# (3) the model's specification.

keys = maes.keys()
df = pd.DataFrame({'name': list(keys), 'mae': maes.values(), 'model': models})
print(df)
```

```python
# The best model
best_model = df["model"][df.mae == min(df.mae)].iloc[0]
# This is the model that has the lowest mae, which is how the best model is selected here.
```
